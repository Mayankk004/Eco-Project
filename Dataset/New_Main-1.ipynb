```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

Reading the Data

```python
df = pd.read_csv(r'ECO221_Project_2025_Final.csv')
df1 = pd.read_csv(r'RF_DistrictWise_ECO221_2025.csv')
df2 = pd.read_csv(r'Salinity_Alkalinity_ECO221_2025.csv')
```

Selecting Only Minor Pulses

```python
MinorPulses_df = df[df['crop'] == 'Minor Pulses']
MinorPulses_df
```

Rainfall summary for the year 2017

```python
df1_summary = df1.groupby(['state', 'district'])['rf'].agg(['sum', 'mean']).reset_index()
df1_summary.rename(columns={'sum': 'total_rainfall', 'mean': 'average_rainfall'}, inplace=True)
df1_summary
```

Mearging the file df and df1

```python
merged_df = MinorPulses_df.merge(df1_summary, left_on=['statename', 'districtname'], right_on=['state', 'district'], how='left')
merged_df.drop(columns=['state', 'district'], inplace=True)
merged_df
```

Merge data with district-level soil quality data

```python
merged_df_Soil = merged_df.merge(df2, left_on='districtname', right_on='District Name', how='left')
merged_df_Soil.rename(columns={'Average Salinity/Alkalinity (Percentage)': 'salinity_alkalinity_percent'}, inplace=True)
merged_df_Soil.drop(columns=['District Name'], inplace=True)
print(merged_df_Soil.head())
```

Data Cleaning

```python
# Step 1: Drop rows with missing area, production, or irrigation (these are essential)
merged_df_Soil = merged_df_Soil.dropna(subset=[
    'area1000hectares', 
    'production1000tonnes', 
    'irrigatedarea1000hectares'
])

# Fill missing average_rainfall and total_rainfall with the state-level mean
merged_df_Soil['average_rainfall'] = merged_df_Soil['average_rainfall'].fillna(
    merged_df_Soil.groupby('statename')['average_rainfall'].transform('mean')
)

merged_df_Soil['total_rainfall'] = merged_df_Soil['total_rainfall'].fillna(
    merged_df_Soil.groupby('statename')['total_rainfall'].transform('mean')
)


# Fill salinity_alkalinity_percent with district mean or overall mean
merged_df_Soil['salinity_alkalinity_percent'] = merged_df_Soil.groupby('districtname')['salinity_alkalinity_percent'].transform(lambda x: x.fillna(x.mean()))
merged_df_Soil['salinity_alkalinity_percent'].fillna(merged_df_Soil['salinity_alkalinity_percent'].mean())

# Capping irrigated area at total cultivated area where it exceeds
merged_df_Soil['irrigatedarea1000hectares'] = merged_df_Soil[['irrigatedarea1000hectares', 'area1000hectares']].min(axis=1)

#Handling missing soil data with state mean
merged_df_Soil['salinity_alkalinity_percent'] = merged_df_Soil.groupby('statename')['salinity_alkalinity_percent'].transform(
    lambda x: x.fillna(x.mean())
)

# Checking for missing values
print("Missing values in each column:")
print(merged_df_Soil.isnull().sum())

```

Handling Outliers

```python
def cap_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    df[column] = df[column].clip(lower, upper)
    return df

# Columns to cap
columns_to_cap = [
    'area1000hectares', 
    'production1000tonnes', 
    'irrigatedarea1000hectares',
    'total_rainfall', 
    'average_rainfall', 
    'salinity_alkalinity_percent',
    'nitrogenconsumptiontonnes', 
    'phosphateconsumptiontonnes', 
    'potashconsumptiontonnes'
]

# Apply capping to each column
for col in columns_to_cap:
    merged_df_Soil = cap_outliers_iqr(merged_df_Soil, col)
```

Make them in Excel file

```python
df1_summary.to_csv(r'MinorPulsesSummary.csv', index=False)
merged_df.to_csv(r'Task1Summary.csv', index=False)
merged_df_Soil.to_csv(r'Task2Summary.csv', index=False)
```
